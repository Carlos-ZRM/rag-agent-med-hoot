```python
import logging
import time
from pathlib import Path

import pandas as pd

from docling.document_converter import DocumentConverter
```

```python
pdf_name = "Allergy_-_2026_-_Zuberbier_-_The_International_Guideline_for_the_Definition__Classification__Diagnosis_and_Management_of.pdf"
logging.basicConfig(level=logging.INFO)

data_folder = "docs/"
input_doc_path = data_folder+ pdf_name

output_dir = Path("scratch")

doc_converter = DocumentConverter()
```

```python
start_time = time.time()
conv_res = doc_converter.convert(input_doc_path)
output_dir.mkdir(parents=True, exist_ok=True)
doc_filename = conv_res.input.file.stem
```

```python
from docling_core.types.doc import TextItem

for item, level in conv_res.document.iterate_items():
    if isinstance(item, TextItem):
        # Document title (top-level heading)
        if item.label == "title":
            print(f"📄 Document title: {item.text}")
        
        # Section headings (h1, h2, h3...)
        if item.label in ("section_heading", "heading"):
            print(f"{'  ' * level}📌 H{level}: {item.text}")
        # Regular text content
        if item.label != "text":
            print("Level:", level, "Label:", item.label, "Text:", item.text[:15])
```
